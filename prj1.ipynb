# **Music recommendation program**


## **loading data**

```python
import pandas as pd
music_data = pd.read_csv('/content/csv file/music.csv')
music_data
```

## **data prep**

### splitting the data into input and output data i.e (X, y)

```python
X = music_data.drop(columns = ['genre'])
X
```

```python
y = music_data ['genre']
y
```

### data split to training and test sets

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
X_train
```

```python
X_test
```

## **decision tree**

### training the model

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(X_train,y_train)

```

### making predictions

```python
Test_predictions = model.predict(X_test)
Train_predictions = model.predict(X_train)
```

```python
Test_predictions
```

```python
Train_predictions
```

### measuring accuracy

```python
from sklearn.metrics import accuracy_score, mean_squared_error, r2_score

accuracy = accuracy_score(y_test, Test_predictions)


Result = pd.DataFrame(['decision tree', accuracy]).transpose()
Result.columns = ["method", "accuracy"]

Result
```

### making persisting model

```python
import joblib

joblib.dump(model, "music_recommender.joblib")

# to call the model
# "variable" = joblib.load("file ko naam.joblib")
```

### visualizing a decision tree

```python
from sklearn import tree
tree.export_graphviz(model, out_file = '/content/music-recommender.dot',
                     feature_names = ["age", "gender"],
                     class_names = sorted(y.unique()),
                     label = "all",
                     rounded = True,
                     filled = True)
```
